```python
import cv2
import numpy as np
import pandas as pd 
import glob
import os
import cv2
from matplotlib import pyplot as plt
#import opencv_jupyter_ui as jcv2
#%matplotlib inline

#The video partition
# SPLITS = ['val', 'test', 'train']
SPLITS = ['train']

#the name of the recording that is to be observed
recordingToSearch = "20190808_171636"

string_start_array = ["_Color", "_Depth"]
string_finish = ".txt"
path = '/mnt/gpid08/datasets/How2Sign/How2Sign_RGB-D/videos/'

#searching the video
for partition in SPLITS:    
    for data_folder in os.listdir(path + partition + "/"):
        for date_folder in os.listdir(path + partition + "/" + data_folder + "/"):
            for recording in os.listdir(path + partition + "/" + data_folder + "/" + date_folder + "/"):
                if recording == recordingToSearch:
                    #loading the video
                    print(f' ---- Video to analyse ---- {recording}', flush = True)
                    for video in os.listdir(path + partition + "/" + data_folder + "/" + date_folder + "/" + recording + "/"):
                        if string_start_array[0] in video:
                            cap = cv2.VideoCapture(path + partition + "/" + data_folder + "/" + date_folder + "/" + recording + "/" + video)
                            cnt = 0
                            while True:
                                ret, frame = cap.read()
                                if ret == False:        break
                                
                                #Print every 20 frames
                                if cnt == 20:
                                    frame = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
                                    plt.imshow(frame)
                                    plt.show()
                                    cnt = 0
                                else:
                                    cnt+=1
                                    
```
